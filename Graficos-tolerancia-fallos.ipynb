```python
from datetime import datetime
import matplotlib.pyplot as plt 
```

```python

x=["Application","Infrastructure","Availability Zone","Region"]

y=[60,0,0,60]

z = [1,1,1,1]

def graficar(x,y,z):
    plt.title('Foundational IT core Services')
    plt.suptitle('Tolerancia de fallos RTO', fontsize=14, fontweight='bold')
    plt.xlabel('Services')
    plt.ylabel('Time (minutes)')
    plt.plot_date(x,z,linestyle ='solid', label="Targeted")
    plt.plot_date(x, y,linestyle ='solid', label="Estimated")
    plt.gcf().set_size_inches(8, 6)
    plt.tight_layout()
    plt.legend()
    plt.show()
    

graficar(x,y,z)
```
